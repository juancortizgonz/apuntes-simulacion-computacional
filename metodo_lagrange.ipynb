# Método de Lagrange

El **método de Lagrange** es una técnica utilizada en interpolación numérica para encontrar un polinomio que pase exactamente por un conjunto dado de puntos. Dado un conjunto de $n+1$ puntos $(x_i, y_i)$ donde $i=0, 1, ..., n$, el polinomio interpolante de Lagrange es un polinomio de grado $n$ que pasa exactamente por todos estos puntos.

## Fórmula del Polinomio de Lagrange

El polinomio de interpolación de Lagrange se define como:

$$
P_n(x)=\sum_{i=0}^{n}{L_i(x)y_i}
$$

Donde $L_i(x)$ son los polinomios básicos de Lagrange definidos como:

$$
L_i(x)=∏_{0 \leq j \leq n ; j \neq i} \frac{x-x_j}{xi-x_j}
$$

## Pasos para construir el polinomio de Lagrange

1. **Determinar los puntos de interpolación:** Tienes que tener un conjunto de puntos $(x_0, y_0), (x_1, y_1), ..., (x_n, y_n)$ que se usarán para construir el polinomio.
2. **Construir los polinomios $L_i(x)$:** Para cada $i$, calcula el polinomio $L_i(x)$. Este polinomio es un producto de fracciones que involucran todos los puntos excepto $x_i$.
3. **Construir el polinomio final $P_n(x)$:** Suma todos los términos de la forma $L_i(x)y_i$ para obtener el polinomio final.

## Ejemplo

Supongamos que tenemos los siguientes puntos de interpolación:

$$
(1, 1), (2, 4), (3, 9)
$$

Queremos encontrar el polinomio que pasa por estos puntos.

1. **Construir los polinomios $L_0(x), L_1(x), L_2(x)$:**
$$
L_0(x)=\frac{(x-2)(x-3)}{(1-2)(1-3)}=\frac{(x-2)(x-3)}{2}  \newline
$$
$$
L_1(x)=\frac{(x-1)(x-3)}{(2-1)(2-3)}=-(x-1)(x-3)
$$
$$  
L_2(x)=\frac{(x-1)(x-2)}{(3-1)(3-2)}=\frac{(x-1)(x-2)}{2}  
$$

2. **Construir el polinomio de Lagrange:**
$$
P_2(x)=L_0(x)y_0+L_1(x)y_1+L_2(x)y_2
$$

Sustituyendo los valores de $y_0=1, y_1=4, y_2=9$:

$$
P_2(x)=(\frac{(x-2)(x-3)}{2})\cdot 1 + (-(x-1)(x-3)) \cdot 4 + (\frac{(x-1)(x-2)}{2}) \cdot 9
$$

Ahora, si expandimos y simplificamos esta expresión, obtenemos un polinomio interpolante de grado 2:

$$
P_2(x)=x^2
$$

Este polinomio pasa exactamente por los puntos dados: $(1, 1), (2, 4), (3, 9)$.
