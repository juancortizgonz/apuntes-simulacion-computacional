# Método de Newton

El **método de Newton** es una técnica de interpolación, pero se basa en una forma diferente de construir el polinomio interpolante. A diferencia de Lagrange, que utiliza un conjunto de productos, el método de Newton utiliza una forma incremental que se puede expandir de manera sucesiva.

## Fórmula del Polinomio de Newton

El polinomio de interpolación de Newton se construye en base a diferencias divididas. El polinomio interpolante de Newton tiene la siguiente forma:

$$
P_n(x)=f(x_0)+(x-x_0) \cdot \Delta f(x_0, x_1)+(x-x_0)(x-x_1) \cdot \Delta^2 f(x_0, x_1, x_2) + ...
$$

Donde las diferencias divididas $\Delta f(x_0, x_1, ..., x_k)$ se calculan de la siguiente manera:

- $\Delta^0 f(x_i)=f(x_i)$
- $\Delta^1 f(x_i, x_{i+1})=\frac{f(x_{i+1})-f(x_i)}{x_{i+1}-x_i}$
- $\Delta^2 f(x_i, x_{i+1}, x_{i+2})=\frac{\Delta^1 f(x_{i+1}, x_i+2)-\Delta^1 f(x_i, x_{i+1})}{x_{i+2}-x_i}$
- Y así sucesivamente.

## Pasos para construir el polinomio de Newton

1. Organizar los puntos de interpolación en una tabla de diferencias divididas.
2. Construir el polinomio de Newton a partir de las diferencias divididas.

## Ejemplo

Consideremos los puntos:

$$
(1, 1), (2, 4), (3, 9)
$$

Queremos encontrar el polinomio que pasa por estos puntos usando el Método de Newton.

1. Organizar los puntos de una tabla de diferencias divididas:

$$
x_i=(1, 2, 3)
$$
$$
f(x_i)=(1, 4, 9)
$$
$$
\Delta^1 f(x_0, x_1) = (empty, 3, 5)
$$
$$
\Delta^2 f(x_0, x_1, x_2) = (empty, empty, 1)
$$

Las diferencias divididas se calculan de la siguiente forma:

- $\Delta^1 f(1, 2)=\frac{4-1}{2-1}=3$
- $\Delta^1(2, 3)=\frac{9-4}{3-2}=5$
- $\Delta^2 f(1, 2, 3)=\frac{5-3}{3-1}=1$

2. Construir el polinomio de Newton:

$$
P_2(x)=f(1)+(x-1) \cdot \Delta^1 f(1, 2) + (x-1)(x-2) \cdot \Delta^2 f(1, 2, 3)
$$

Sustituyendo los valores:

$$
P_2(x)=1+(x-1) \cdot 3 + (x-1)(x-2) \cdot 1
$$

Simplificando y luego expandiendo tenemos:

$$
P_2(x)=x^2
$$

Este es el mismo polinomio que se obtuvo usando el método de Lagrange, que pasa por los puntos específicados.
